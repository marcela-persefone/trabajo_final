```python
import pandas as pd
import matplotlib.pyplot as plt
```

##### Esta leyendo el archivo y creando una variable dataframe llamada df

```python
df = pd.read_csv('BASE.csv')
```

##### vamos  aplicar One Hot Encoding, prefix='Gender' Agrega un prefijo a cada nueva columna para que el nombre quede claro

```python
gender_encoded = pd.get_dummies(df['Gender'], prefix='Gender')
```

#### vamos a eliminar la variable Gender, luego vamos hacer una nueva tabla llamada df_encoded 

```python
df_encoded = df.drop('Gender', axis=1)
df_encoded = pd.concat([df_encoded, gender_encoded.astype(int)], axis=1)
```

#### Observamos como quedaron nuestras varaibles

```python

print(df_encoded.head())
```

### se sube una nueva base

```python
df_encoded.to_csv('base_codificada_gender.csv', index=False)
```

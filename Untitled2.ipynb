```python
import os
print(os.getcwd())

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snb

df = pd.read_csv('base1.csv')
df.head()
df.info()
df.describe(include='all')
df["Gender_Dummy"] = (df["Gender"] == "Male").astype(int)
df[["Gender", "Gender_Dummy"]].head()


```

```python
df["Purchase_Amount"] = df["Purchase_Amount"].replace('[\$,]', '', regex=True).astype(float)
df["Purchase_Amount"].head()
df["Purchase_Amount"].isna().sum()
df[df["Purchase_Amount"] <= 0]

```

```python
plt.hist(df["Purchase_Amount"], bins=30)
plt.title("Distribución de Purchase Amount")
plt.xlabel("Monto")
plt.ylabel("Frecuencia")
plt.show()
variables = [
    "Frequency_of_Purchase",
    "Purchase_Category",
    "Age",
    "Gender",
    "Purchase_Amount",
    "Social_Media_Influence",
    "Engagement_with_Ads",
    "Device_Used_for_Shopping"
]
```

```python
df_model = df[variables].copy()

# Agrupar categorías poco frecuentes de 'Purchase_Category' en 'Other'
category_counts = df_model["Purchase_Category"].value_counts()
rare_threshold = 20
rare_categories = category_counts[category_counts < rare_threshold].index

df_model["Purchase_Category_Grouped"] = df_model["Purchase_Category"].apply(
    lambda x: "Other" if x in rare_categories else x
)

# (Opcional) Ver nueva distribución agrupada
print(df_model["Purchase_Category_Grouped"].value_counts())
```

```python
for col in ["Social_Media_Influence", "Engagement_with_Ads"]:
    df[col].fillna(df[col].mode()[0], inplace=True)
df[["Social_Media_Influence", "Engagement_with_Ads"]].isnull().sum()
```

```python
df_model = pd.get_dummies(df_model, columns=[
    "Frequency_of_Purchase",
    "Purchase_Category",
    "Social_Media_Influence",
    "Engagement_with_Ads"
], drop_first=True)
```
